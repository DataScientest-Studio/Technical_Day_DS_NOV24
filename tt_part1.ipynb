# Code et algorithmes sur Pyhton

```python
# Fonction qui vérifie si une chaîne de caractères (mot ou phrase sans accent ni ponctuation) est un palindrome

def palindrome_complex(string):
    string = string.replace(' ', '').lower()
    if string == string[::-1]:
        return True 
    else:
        return False
    
print(palindrome_complex('La mariee ira mal'))
print(palindrome_complex('La mariee ira bien'))
```

```python
# Fonction qui renvoie la moyenne des chiffres composant un nombre

def moyenne_chiffres(nombre):
    chiffres = [int(chiffre) for chiffre in str(abs(nombre))]

    somme = sum(chiffres)

    return somme / len(chiffres)

moyenne_chiffres(1234589011)
```

```python
# Fonction qui renvoie tous les nombres premiers d'une liste

def nombres_premiers(liste):
    liste_premiers = []

    for n in liste:
        if n < 2:
            continue

        is_premier = True
        for i in range(2, int(n**0.5) + 1):  
            if n % i == 0:
                is_premier = False
                break 

        if is_premier:
            liste_premiers.append(n)

    return liste_premiers

nombres_premiers(list(range(1,30)))
```

```python
# Import des librairies

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
```

```python
# Chargement du fichier + visualisation du dataframe

df = pd.read_csv('weight_change_dataset.csv')
df.head()
```

```python
# Affichez les information de l'ensemble des hommes ayant perdu du poids

df.loc[(df['Gender'] == 'M') & (df['Weight Change (lbs)'] < 0)]
```

```python
# Affichez les informations de la 16eme personne la plus agée

df.sort_values(by = 'Age', ascending=False).iloc[15]
```

```python
# Comptez le nombre de participants qui ont plus de 30 ans et une qualité de sommeil 'excellent' ou 'good'

df_bis = df.loc[(df['Age'] > 30) & ((df['Sleep Quality'] == 'Excellent') | (df['Sleep Quality'] == 'Good'))]
```

```python
# Affichez la courbe de densité de la variable `Age` en différenciant selon le genre

plt.figure(figsize=(10, 6))
sns.kdeplot(x = 'Age', hue = 'Gender', data = df) 

plt.title("Courbe de densité de l'âge par genre")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
# La difficulté à perdre du poids varie-t-elle avec l'âge, la qualité du sommeil et le niveau de stress d'après nos données ? Utilisez un graphique pour explorer ces relations.

plt.figure(figsize=(12, 8))
sns.scatterplot(x = 'Age', y = 'Weight Change (lbs)', hue = 'Sleep Quality', size = 'Stress Level', data = df)

plt.title("Relation entre l'âge, la qualité du sommeil, le niveau de stress et la variation de poids")
plt.xlabel("Âge")
plt.ylabel("Variation de poids (lbs)")
plt.show()
```
